```python
def perceptron(x,y):

    # insert 1 in x
    x= np.insert(x,0,1,axis=1)
    # intial weights
    weights= np.ones(x.shape[1])
    lr=0.1
    
    for i in range(1000):
        j=np.random.randint(0,100)
        y_hat = step(np.dot(x[j],weights))
        # y_hat is y predicted
        # update the weights
        weights = weights + lr*(y[j]-y_hat)*x[j]
        
    return weights[0], weights[1:]

# weights[0] is weights[w0] ,w0 is intercept
# weights[1:] = weights[w1,w2,w3,---],   it is coefficients.       
```

```python
intercept,coef = perceptron(x,y)
```

```python
m = -( coef[0]/coef[1])
b = -( intercept/coef[1])
```
